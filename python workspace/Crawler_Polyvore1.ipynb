# Polyvore Crawler for Today Outfit Links

```python
import requests
import json
import sqlite3
from lxml import html
from bs4 import BeautifulSoup
from datetime import date,datetime
```

```python
# Headers

head = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/43.0.2357.65 Safari/537.36'}
```

```python
# http://www.polyvore.com/

# Today Outfit (Category: Women's Fashion)
                    
res = requests.get('https://www.polyvore.com/cgi/search.sets?category=fashion&date=day&item_count.from=4&item_count.to=10', headers = head)
soup = BeautifulSoup(res.text, "html.parser")
```

```python
# Today's Outfit Page1

outfit_link = []
for title in soup.select('.layout_n.grid.grid_6')[0]:
    # All links in page1
    for link in title.select('div.title'):
        outfit_link.append('https://www.polyvore.com/' + link.select('a')[0]['href'][3:])

# One page stores 30 links
print len(outfit_link)
for link in outfit_link:
    print link
```

```python
# Today's Outfit Page2

res = requests.get('https://www.polyvore.com/cgi/search.sets?.in=json&.out=jsonx&request={"item_count.to":"10","date":"day","category":"fashion","item_count.from":"4","page":2,".passback":{"next_token":{"limit":"30","start":30}}}', headers = head)

data = json.loads(res.text)
html = data['result']['html']
soup = BeautifulSoup(html, "html.parser")

for title in soup.select('.layout_n.grid.grid_6')[0]:
    #All links in page2
    for link in title.select('div.title'):
        outfit_link.append('https://www.polyvore.com/' + link.select('a')[0]['href'][3:])

print len(outfit_link)
for link in outfit_link:
    print link
```

```python
# Today's Outfit Page3 ~ PageXX

last = 30 # Last page number 

for i in xrange(last-2): # i = 0 ~ last-2-1
    j = 60+(i*30) # Start from 60
    res = requests.get('https://www.polyvore.com/cgi/search.sets?.in=json&.out=jsonx&request={"item_count.to":"10","date":"day","category":"fashion","item_count.from":"4","page":null,".passback":{"next_token":{"limit":"30","start":' + str(j) +'}}}', headers = head)
    
    data = json.loads(res.text)
    html = data['result']['html']
    soup = BeautifulSoup(html, "html.parser")

    for title in soup.select('.layout_n.grid.grid_6')[0]:
        #All links in this pages
        for link in title.select('div.title'):
            outfit_link.append('https://www.polyvore.com/' + link.select('a')[0]['href'][3:])

print len(outfit_link)
for link in outfit_link:
    print link
```

```python
# Write outfit links into a txt file

day = datetime.now().strftime("%Y%m%d") 

with open('outfit_links/Today_Outfit_Links_'+ day +'.txt', 'w') as f:
    for link in outfit_link:
        f.write(link + '\n')
```
