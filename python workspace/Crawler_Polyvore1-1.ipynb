## Polyvore Crawler for The last 3 months Outfit Links

```python
import requests
import json
import sqlite3
from lxml import html
from bs4 import BeautifulSoup
from datetime import date,datetime
```

```python
# Headers

head = {'User-Agent':'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/43.0.2357.65 Safari/537.36'}
```

```python
# http://www.polyvore.com/

# The last 3 months Outfit (Category: Women's Fashion)

res = requests.get('https://www.polyvore.com/cgi/search.sets?category=fashion&date=3m&item_count.from=4&item_count.to=10', headers = head)
soup = BeautifulSoup(res.text, "html.parser")
```

```python
# The last 3 months Outfit Page1

outfit_link = []
for title in soup.select('.layout_n.grid.grid_6')[0]:
    # All links in page1
    for link in title.select('div.title'):
        outfit_link.append('https://www.polyvore.com/' + link.select('a')[0]['href'][3:])

# One page stores 30 links
print len(outfit_link)
for link in outfit_link:
    print link
```

```python
# The last 3 months Outfit Page2 ~ PageXX

last = 30 # Last page number 

for i in xrange(last-2): # i = 0 ~ last-2-1
    j = 60+(i*30) # Start from 60
    res = requests.get('https://www.polyvore.com/cgi/search.sets?.in=json&.out=jsonx&request=%7B%22item_count.to%22%3A%2210%22%2C%22date%22%3A%223m%22%2C%22category%22%3A%22fashion%22%2C%22item_count.from%22%3A%224%22%2C%22page%22%3Anull%2C%22.passback%22%3A%7B%22next_token%22%3A%7B%22limit%22%3A%2230%22%2C%22start%22%3A'+ str(j) +'%7D%7D%7D', headers = head)
    
    data = json.loads(res.text)
    html = data['result']['html']
    soup = BeautifulSoup(html, "html.parser")

    for title in soup.select('.layout_n.grid.grid_6')[0]:
        #All links in this pages
        for link in title.select('div.title'):
            outfit_link.append('https://www.polyvore.com/' + link.select('a')[0]['href'][3:])

print len(outfit_link)
for link in outfit_link:
    print link
```

```python
# Write outfit links into a txt file

day = datetime.now().strftime("%Y%m%d") 

with open('outfit_links/3Months_Outfit_Links_'+ day +'.txt', 'w') as f:
    for link in outfit_link:
        f.write(link + '\n')
```

```python
# Remove duplicate lines in the file

lines = open('outfit_links/3Months_Outfit_Links_'+ day +'.txt', 'r').readlines()

lines_set = set(lines)

output = open('outfit_links/3Months_Outfit_Links_'+ day +'.txt', 'w')

for line in lines_set:
    output.write(line)
```
