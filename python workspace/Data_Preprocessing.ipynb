```python
# Read outfits data

outfits_df = pd.read_csv("outfits.csv")
outfits_df = outfits_df.drop(["link", "title", "image", "author", "views", "description", "similar1", "similar2", "similar3", "similar4", "similar5"], axis=1)
outfits_df = outfits_df.rename(columns={"likes": "outfit_likes"})
outfits_df = outfits_df.sort_values('outfit_id').reset_index(drop=True)

outfits_df
```

```python
# Read items data

items_df = pd.read_csv("items.csv")
items_df = items_df.drop(["link", "title", "description", "image", "category1_id", "category1_name", "category2_id", "category2_name", "category3_id", "category3_name", "category4_id", "category4_name", "brand", "price", "keyword1", "keyword2", "keyword3", "keyword4", "keyword5", "keyword6"], axis=1)
items_df = items_df.rename(columns={"likes": "item_likes"})

items_df
```

```python
# Read item category data

item_category_df = pd.read_csv("item_category.csv")
item_category_df = item_category_df.rename(columns={"category1": "item_category1", "category2": "item_category2", "category3": "item_category3"})

item_category_df
```

```python
# Read outfit_item links data

outfit_item_df = pd.read_csv("outfit_item.csv")
outfit_item_df = outfit_item_df.drop(["id"], axis=1)
outfit_item_df = outfit_item_df.sort_values('outfit_id').reset_index(drop=True)

outfit_item_df
```

```python
# Merge outfits and items

outfit_item_df = pd.merge(outfit_item_df, outfits_df)
outfit_item_df = pd.merge(outfit_item_df, items_df)
outfit_item_df = pd.merge(outfit_item_df, item_category_df)
outfit_item_df = outfit_item_df.sort_values('outfit_id').reset_index(drop=True)
# outfit_item_df = outfit_item_df[outfit_item_df.create_date == 20171201]

outfit_item_df
```

```python
# Concat two item image to an image
def concatImage(date, outfit, likes, item1, item2):
    if likes >= 50:
        category = 1
    else:
        category = 0
        
    input1 = 'item_images/'+str(date)+'/'+str(date)+'-'+str(outfit)+'-'+str(item1)+'.jpg'
    input2 = 'item_images/'+str(date)+'/'+str(date)+'-'+str(outfit)+'-'+str(item2)+'.jpg'
#     output = 'data/Outfits/Train/'+str(category)+'/'+str(outfit)+'-'+str(item1)+'-'+str(item2)+'.jpg'
    output = 'data/Outfits/Test/'+str(outfit)+'-'+str(item1)+'-'+str(item2)+'.jpg'
    
    images = map(Image.open, [input1, input2])
    widths, heights = zip(*(i.size for i in images))

    total_width = sum(widths)
    max_height = max(heights)

    new_im = Image.new('RGB', (total_width, max_height))

    x_offset = 0
    for im in images:
        new_im.paste(im, (x_offset,0))
        x_offset += im.size[0]

    new_im.save(output)
```
